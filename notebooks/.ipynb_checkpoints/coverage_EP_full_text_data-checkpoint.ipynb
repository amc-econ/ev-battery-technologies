## Introduction

According to its offical documentation, "EP full-text data for text analytics contains all EP publications, from 1978 until the end of January 2019". $^{1}$ However, many important patents are filled under different juridictions in order to maximise the geographical coverage of the patent collection. **For this reason, this notebook looks at the *implicit* coverage of the dataset.** This approach follows an exercise undertaken by ***C. Verluise***, but only for the subset of data of interest (wind technologies and later EV battery technologies). $^{2}$

## 2. Coverage of the EP office versus worldwide patents

### 2.1. With the entire dataset

The computation below confirms that non-EU patents can be also be covered in the EU by another member of the DOCDB patent family!

```python
def assess_coverage(data):
    
    data = data.copy()
    
    top_authorities = data.appln_auth.value_counts().to_frame().reset_index()['index'].unique().tolist()[0:15]

    for authority in top_authorities:
        us_patent_fam_ids = data[data.appln_auth == authority].docdb_family_id.unique().tolist()
        df = data[data.docdb_family_id.isin(us_patent_fam_ids)]
        print(authority)
        df = df.appln_auth.value_counts().to_frame()
        proportion_in_EP = round(df.at['EP','appln_auth']/df.at[authority,'appln_auth']*100,1)
        print('Proportion of EP patents vs local patents: {}%'.format(proportion_in_EP))
        display(df)
        print('---')
```

```python
assess_coverage(data)
```

### 2.2 Selection of patents of interest

Let see if this proportion gets higher when we select only the highest cited patents!

```python
import math
import warnings
warnings.filterwarnings("ignore")
```

```python
def _get_DOCDB_fam_cites_per_year(data):
        """Adding a variable to keep track of yearly citations by patent family"""
        
        data = data.copy()
        
        # Unpacking some variables for clarity
        citations_fam_by_year = 'citations_fam_per_year'
        citations_docdb_fam = 'nb_citing_docdb_fam'
        earliest_filling_year = 'earliest_filing_year'
        ref_year = 2018
        
        # add a variable for the earliest filling year to Pandas datetime format
        #data.earliest_filing_date = pd.to_datetime(data.earliest_filing_date, format = '%Y-%m-%d')
        
        # for each patent get the earliest filling date
        #data.earliest_filling_year = data.earliest_filing_date.dt.year
        
        # Compute the metric
        data[citations_fam_by_year] = data[citations_docdb_fam]/(ref_year-data[earliest_filling_year])
        
        return data
```

```python
data = _get_DOCDB_fam_cites_per_year(data)
```

```python
def _select_one_patent_per_family(data):
        """In order to select only patent of interest, as well as
        saving computationnal power, we select only the earliest patent by
        family"""
        
        data = data.copy()
        
        # Filtering 
        data.sort_values(by = 'earliest_filing_date',inplace = True)
        data.drop_duplicates(subset = ['docdb_family_id'],keep = 'first',inplace = True)
        
        return data
```

```python
def _select_breakthrough_families(data, percentage_top_families):
        """Filtering the data to keep only breakthrough patents"""
        
        data = data.copy()
        print(len(data))
        
        # keep only the most recent patent for each family
        data_filtered = _select_one_patent_per_family(data)
        print(len(data_filtered))
        
        # for each year, keep only the most cited families (top X%)
        filtered_df = pd.DataFrame()
        for year in data_filtered['earliest_filing_date'].unique().tolist():
            df_year = data_filtered[data_filtered['earliest_filing_date'] == year]
            df_year.sort_values(by = 'nb_citing_docdb_fam',ascending = False,inplace = True)
            
            X = percentage_top_families
            nb_top_patent_given_year = int(math.ceil(X*len(df_year))) # Needs rounding up
            df_year = df_year.head(nb_top_patent_given_year)
            filtered_df = pd.concat([filtered_df, df_year])
            
        # store the families ids 
        fam_ids = filtered_df['docdb_family_id'].unique().tolist()
        
        # filter data with ths family ids
        condition = data['docdb_family_id'].isin(fam_ids)
        
        return data[condition]
```

```python
data_top_patents = _select_breakthrough_families(data = data, percentage_top_families = 0.1)
```

> We see that the coverage is very good for the top 10% of patents for all juridictions except the US.

```python
assess_coverage(data_top_patents)
```

----------------

```python
nb = round(len(data_top_patents['docdb_family_id'].unique()) / len(data['docdb_family_id'].unique()) * 100,1)
print('The number of docdb_families in the filtered data is {}% of the initial data'.format(nb))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(15,5))
fig.suptitle('Distribution of family citations per year\n(left: unfiltered dataset, right: dataset containg top patents)')
sns.distplot(data['citations_fam_per_year'], ax = ax1)
sns.distplot(data_top_patents['citations_fam_per_year'], ax = ax2);
```
